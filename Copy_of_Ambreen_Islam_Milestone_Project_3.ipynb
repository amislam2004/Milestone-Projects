##Task 6: Modeling

Use the following code block to build your ML model.  

```python
# Split data to Training data and Test data 
# we are creating the four arrays and we will using the train test split function. 
# X function will be split into two arrays and one array will be the Training data and other one will be Test data
# Y function will be split into two arrays and one array will be the Training data and other one will be Test data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# 
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Data Standardization

scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)

X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

```python
from matplotlib.cbook import flatten
from pandas.core.reshape.melt import lreshape
# Algorithm Applications
from sklearn.metrics import f1_score, precision_score, recall_score
list_met = []
list_accuracy = []

# Applying Logistic Regression

classifier = LogisticRegression(C=0.4, max_iter= 1000, solver = 'liblinear')
lr = classifier.fit(X_train,Y_train)
# Prediction
Y_predict= classifier.predict(X_test)
#Accuracy
accuracy_LR = accuracy_score(Y_test, Y_predict)

# Applying Decision Tree

classifier2 = DecisionTreeClassifier(random_state=14)
dt = classifier2.fit(X_train, Y_train)
# Prediction
Y_predict2= classifier2.predict(X_test)
#Accuracy
accuracy_DT = accuracy_score(Y_test, Y_predict2)

# Apply Random Forest Criteria = information gain

classifier3 = RandomForestClassifier(random_state=14)
rfi = classifier3.fit(X_train, Y_train)
# Prediction
Y_predict3= classifier3.predict(X_test)
#Accuracy
accuracy_RFI = accuracy_score(Y_test, Y_predict3)

# Applying Random Forest Crieteria = entropy

classifier4 = RandomForestClassifier(criterion='entropy')
rfe = classifier4.fit(X_train, Y_train)
# Prediction
Y_predict4= classifier4.predict(X_test)
#Accuracy
accuracy_RFE = accuracy_score(Y_test, Y_predict4)

# Apply SVM

# Model Training
# Support Vector Machine Model
model_svm = SVC(cache_size=100)
# training the SVM model with training data
svm = model_svm.fit(X_train, Y_train)
# Prediction
Y_predict5= model_svm.predict(X_test)
# Accuracy
accuracy_svc = accuracy_score(Y_test, Y_predict5)

# Apply KNN

model_knn3 = KNeighborsClassifier(n_neighbors=3)
knn = model_knn3.fit(X_train, Y_train)
# Prediction 
pred_knn3 = model_knn3.predict(X_test)
# Accuracy
accuracy_SVM = accuracy_score(Y_test, pred_knn3)

# Apply Gaussian Naive Bayes

gnb = GaussianNB()
gnb = gnb.fit(X_train, Y_train)
#Predicting Test Set
pred_gnb = gnb.predict(X_test)
# Accuracy
accuracy_GNB = accuracy_score(Y_test, pred_gnb)

# Apply Bernoulli Naive Bayes

model = BernoulliNB()
bnb = model.fit(X_train, Y_train)
# Predicting Test set
pred_bnb = model.predict(X_test)
# Accuracy
accuracy_BNB = accuracy_score(Y_test, pred_bnb)

# Combining all the modles using voting classifier

evc = VotingClassifier(estimators=[('lr', lr), ('rfi', rfi), ('rfe', rfe), ('DT', dt), 
                                   ('svm', svm),('knn', knn), ('gnb', gnb), ('bnb', bnb)], voting = 'hard',
                                    flatten_transform = True)
model_evc = evc.fit(X_train, Y_train)
# Predicting Test Set
pred_evc = evc.predict(X_test)
# Accuracy
accuracy_evc = accuracy_score(Y_test, pred_gnb)

list1 = ['Logistic Regression', 'Decision Tree', 'Random Forest (info gain)', 'Random Forest (Entropy)', 'SVM', 'KNN', 'gnb', 'bnb', 'Voting Classifier']
list2 = [accuracy_LR, accuracy_DT, accuracy_RFI, accuracy_RFE, accuracy_svc, accuracy_SVM, accuracy_GNB, accuracy_BNB, accuracy_evc]
list3 = [classifier, classifier2, classifier3, classifier4, model_svm, model_knn3, gnb, model]

parkinsons_data_Accuracy = pd.DataFrame({'Method Used':list1, 'Accuracy': list2})
print(parkinsons_data_Accuracy)

chart = sns.barplot(x= 'Method Used', y='Accuracy', data= parkinsons_data_Accuracy)
chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
print(chart)







```

```python

# Visualization of the performance with ROC 
# Calculate the fpr and tpr for all threshold of the classification
def plot_roc(model, X_test, Y_test):
    probabilities = model.predict_proba(np.array(X_test))
    predictions = probabilities
    fpr, tpr, threshold = roc_curve(Y_test, predictions[:,1])
    roc_auc = auc(fpr, tpr)

    plt.title('Recevier Perating Characteristics')
    plt.plot(fpr, tpr, 'b', label='Auc = %0.2f' % roc_auc)
    plt.legend(loc='lower right')
    plt.plot([0,1], [0,1],'r--')
    plt.xlim([0,1])
    plt.ylim([0,1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
```
